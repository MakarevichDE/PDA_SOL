10) Создать матрицу случайных чисел размера 10 на 10 и отнять среднее по строкам из каждого элемента строки

```python
d=np.random.random((10,10))
print('Исходный массив: \n',d)
e=d.mean(axis=1, keepdims=True)
print('Среднее по строкам: \n',e)
d -=e
print('Результат: \n',d)
```

11) Создать матрицу случайных чисел размера 10 на 10 и отнять медиану по столбцам из каждого элемента столбца  

```python
d=np.random.random((10,10))
print('Исходный массив: \n',d)
e=np.median(d,axis=0, keepdims=True)
print('Медиана по столбцам: \n',e)
d -=e
print('Результат: \n',d)
```

12) Создать матрицу случайных чисел 5x5. Добавить к ней столбцы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
d=np.random.random((5,5))
print('Исходный массив: \n',d)
e=d.max(axis=1, keepdims=True)
f=d.min(axis=1,keepdims=True)
print('Результат: \n',np.hstack((d,e,f)))
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу этого произведения.  

```python
z=np.arange(1,52)
print(z)
h=z[z % 2 != 0]
h=h[0:25].reshape(5,5)
print(h)
i=z[z % 2 == 0].reshape(5,5)
print(i)
o=h.dot(i)
print('Произведение матриц: \n',o)
print('Обратная матрица не может быть построена, потому что определитель равен 0: ',np.linalg.det(o))
```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
k = np.zeros((8,8),dtype=int)
k[1::2,::2] = 1 #нечетные элементы каждой второй строки
k[::2,1::2] = 1 #нечетные элементы каждого второго столбца
print(k)
```
