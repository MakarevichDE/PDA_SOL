3.3. Напишите программу на Python для создания круговой
диаграммы с названием популярности языков программирования. Сделайте несколько клиньев круговой диаграммы.

```python
sns.set_style('darkgrid')
sns.set(font_scale=1.3)
hr = pd.read_csv('ibm-hr-analytics-attrition-dataset.zip')
```

```python
hr.head()
```

```python
hr.dtypes
```

1)	построить	несколько	графиков	на	одном	поле;

```python
sns.relplot(
    x="Age",
    y="JobLevel",
    data=hr,
    color='blue', height=4,
    kind='line', # строит линейный график
    col='Attrition'); # рисует график отношений двух классов по 'Выгоранию'.
```

Можно проследить, что у людей с выгоранием разброс по карьерной лестнице довольно заметный - их положение в компании довольно нестабильное. В то время как люди без выгорания чаще идут вверх по карьерной лестнице.

2)	построить	диаграммы	для	категориальных	данных;

```python
sns.catplot(
    x='JobLevel',
    y='MonthlyIncome',
    data=hr,
    palette='bright',
    kind='bar',
    hue='Gender');
```

Можно заметить, что заработная плата у мужчин и женщин практически одинаковая в компании.

```python
sns.catplot(
    x='Attrition',
    y='MonthlyIncome',
    data=hr,
    palette='bright',
    kind='bar',
    col='Department');
```

Вывод: люди с меньшей зарплатой больше подвержены выгоранию.

3)	построить	столбчатые	и	круговые	диаграммы;

```python
df = hr.groupby('EnvironmentSatisfaction').size()

# Make the plot with pandas
df.plot(kind='pie', subplots=True, figsize=(8, 8), autopct='%1.1f%%')
plt.title("Доля удовлетворенных условиями труда (1-Минимально, 4-Максимально)")
```

Большинство людей нейтрально относится к рабочей среде.

```python
sns.barplot(
    x='Education',
    y='MonthlyIncome',
    data=hr,
    color='blue')
```

Вывод: чем выше ученая степень, тем больше зарплата.

4)	реализовать	цветовую	сетку	данных;

```python
sns.set(font_scale=0.5)
plt.figure(figsize=(15,12))
sns.heatmap(
    hr.corr(),
    cmap='RdBu_r', # задаёт цветовую схему
    annot=True, # рисует значения внутри ячеек
    vmin=-1, vmax=1); # указывает начало цветовых кодов от -1 до 1.
```

С помощью тепловой карты можно установить корелляцию между признаками, а именно:
1. Возраст и Всего проработанных лет. Можно говорить, что чем больше человек проработал лет, тем выше он по карьерной лестнице.
2. Уровень работы и Ежемесячная зарплата. Корелляция почти единица, что логично, чем выше человек по карьерной лестнице, тем выше его зарплата.
